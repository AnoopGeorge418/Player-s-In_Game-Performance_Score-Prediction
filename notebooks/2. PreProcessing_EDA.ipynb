# Splitting Dataset for Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_cleaned.drop('win_ranking_in_%', axis = 1)
```

```python
X
```

```python
y = df_cleaned.iloc[:, -2]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# Standardizing the dataset

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Using Lasso - reduces to 0 which feature has low importance

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha=0.01)
lasso.fit(X_train_scaled, y_train)
```

```python
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importance})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
importance_df
```

### Using Recursive Feature Elimination(RFE)

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(X_train_scaled, y_train)

ranking = rfe.ranking_
selected_features = X.columns[rfe.support_]
print("Selected features:", selected_features)
```

# Model Training and Scaling

```python
top_features = ['boosts', 'players_knocked', 'kill_rank', 'revives', 'rankPoints', 'matchDuration', 'longestKill', 'kills', 'winPoints', 'walkDistance']
```

```python
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]
```

```python
X_train_top
```

```python
X_train_top_scaled = scaler.fit_transform(X_train_top)
X_test_top_scaled = scaler.transform(X_test_top)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression()
model.fit(X_train_top_scaled, y_train)

# Make predictions
# y_pred_train = model.predict(X_train_top_scaled)
y_pred_test = model.predict(X_test_top_scaled)

# Evaluate the model
# train_mse = mean_squared_error(y_train, y_pred_train)
test_mse = mean_squared_error(y_test, y_pred_test)
# train_r2 = r2_score(y_train, y_pred_train)
test_r2 = r2_score(y_test, y_pred_test)

# print(f"Training MSE: {train_mse}")
print(f"Testing MSE: {test_mse}")
# print(f"Training R^2: {train_r2}")
print(f"Testing R^2: {test_r2}")
```

# Compare with All Features

```python
# Train the model with all features for comparison
model_all = LinearRegression()
model_all.fit(X_train_scaled, y_train)

# Make predictions
y_pred_train_all = model_all.predict(X_train_scaled)
y_pred_test_all = model_all.predict(X_test_scaled)

# Evaluate the model
train_mse_all = mean_squared_error(y_train, y_pred_train_all)
test_mse_all = mean_squared_error(y_test, y_pred_test_all)
train_r2_all = r2_score(y_train, y_pred_train_all)
test_r2_all = r2_score(y_test, y_pred_test_all)

print(f"Training MSE (All Features): {train_mse_all}")
print(f"Testing MSE (All Features): {test_mse_all}")
print(f"Training R^2 (All Features): {train_r2_all}")
print(f"Testing R^2 (All Features): {test_r2_all}")
```
